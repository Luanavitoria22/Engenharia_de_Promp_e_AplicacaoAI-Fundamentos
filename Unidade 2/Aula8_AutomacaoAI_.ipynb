```python
import os
import shutil
import tempfile

# Dicionário que mapeia extensões de arquivo para nomes de pastas
# Você pode adicionar mais tipos conforme necessário
TIPOS_DE_ARQUIVOS = {
    '.pdf': 'Documentos PDF',
    '.doc': 'Documentos Word',
    '.docx': 'Documentos Word',
    '.xls': 'Planilhas Excel',
    '.xlsx': 'Planilhas Excel',
    '.ppt': 'Apresentações PowerPoint',
    '.pptx': 'Apresentações PowerPoint',
    '.png': 'Imagens PNG',
    '.jpg': 'Imagens JPG',
    '.jpeg': 'Imagens JPEG',
    '.gif': 'Imagens GIF',
    '.txt': 'Arquivos de Texto',
    '.zip': 'Arquivos Compactados',
    '.rar': 'Arquivos Compactados',
    '.mp3': 'Músicas',
    '.wav': 'Músicas',
    '.mp4': 'Vídeos',
    '.avi': 'Vídeos',
    '.mov': 'Vídeos',
    '.py': 'Scripts Python',
    '.ipynb': 'Notebooks Jupyter'
}

def organizar_arquivos(diretorio_origem='.'):
    """
    Organiza arquivos em pastas baseadas em suas extensões no diretório especificado.
    Cria pastas para tipos de documentos comuns (.pdf, .doc, .xls, .png, etc.)
    e move os arquivos para elas. Funciona sem permissões de administrador
    desde que o diretorio_origem seja acessível ao usuário.
    """

    print(f"\nIniciando organização de arquivos em: {os.path.abspath(diretorio_origem)}")

    # Itera sobre todos os itens no diretório de origem
    for nome_arquivo in os.listdir(diretorio_origem):
        caminho_completo_arquivo = os.path.join(diretorio_origem, nome_arquivo)

        # Ignora diretórios
        if os.path.isdir(caminho_completo_arquivo):
            continue

        # Obtém a extensão do arquivo em minúsculas
        _, extensao = os.path.splitext(nome_arquivo)
        extensao = extensao.lower()

        # Verifica se a extensão está no nosso dicionário de tipos
        if extensao in TIPOS_DE_ARQUIVOS:
            nome_pasta = TIPOS_DE_ARQUIVOS[extensao]
        else:
            # Se não for um tipo conhecido, coloca em uma pasta 'Outros'
            nome_pasta = 'Outros'

        # Define o caminho completo para a pasta de destino
        caminho_pasta_destino = os.path.join(diretorio_origem, nome_pasta)

        # Cria a pasta de destino se ela não existir
        if not os.path.exists(caminho_pasta_destino):
            os.makedirs(caminho_pasta_destino)
            print(f"Pasta '{nome_pasta}' criada.")

        # Define o caminho completo do arquivo no destino
        caminho_destino_arquivo = os.path.join(caminho_pasta_destino, nome_arquivo)

        try:
            # Move o arquivo para a pasta de destino
            shutil.move(caminho_completo_arquivo, caminho_destino_arquivo)
            print(f"Arquivo '{nome_arquivo}' movido para '{nome_pasta}'.")
        except Exception as e:
            print(f"Erro ao mover o arquivo '{nome_arquivo}': {e}")

    print("Organização de arquivos concluída!")


def criar_arquivos_teste(diretorio_origem, tipos_de_arquivos_dict, num_arquivos_por_extensao=3):
    """
    Cria arquivos de teste no diretório especificado para simular a organização.
    """
    print(f"\nCriando {num_arquivos_por_extensao} arquivos de teste para cada extensão em: {diretorio_origem}")
    for extensao in tipos_de_arquivos_dict.keys():
        for i in range(1, num_arquivos_por_extensao + 1):
            nome_arquivo = f"arquivo_teste_{i}{extensao}"
            caminho_arquivo = os.path.join(diretorio_origem, nome_arquivo)
            with open(caminho_arquivo, 'w') as f:
                f.write(f"Conteúdo de teste para {nome_arquivo}")
            print(f"Criado: {nome_arquivo}")

# --- EXECUTAR O SCRIPT --- Com simulação de arquivos de teste
print("O diretório de trabalho atual é: " + os.getcwd())

temp_dir = None
try:
    # Cria um diretório temporário para a simulação
    temp_dir = tempfile.mkdtemp()
    print(f"Diretório temporário para simulação criado: {temp_dir}")

    # Cria arquivos de teste no diretório temporário
    criar_arquivos_teste(temp_dir, TIPOS_DE_ARQUIVOS, num_arquivos_por_extensao=3)

    # Executa a organização dos arquivos no diretório temporário
    organizar_arquivos(temp_dir)

except Exception as e:
    print(f"Ocorreu um erro durante a simulação: {e}")
finally:
    # Limpa o diretório temporário
    if temp_dir and os.path.exists(temp_dir):
        shutil.rmtree(temp_dir)
        print(f"\nDiretório temporário '{temp_dir}' e seu conteúdo foram removidos.")
```

```python
import requests
import json

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço.
    Retorna o endereço formatado ou uma mensagem de erro.
    """
    # Remove caracteres não numéricos do CEP
    cep = ''.join(filter(str.isdigit, cep))

    # Validação básica do CEP
    if len(cep) != 8:
        return "Erro: O CEP deve conter exatamente 8 dígitos numéricos."

    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Lança uma exceção para status de erro (4xx ou 5xx)
        data = response.json()

        if "erro" in data and data["erro"]:
            return f"Erro: CEP '{cep}' não encontrado."
        else:
            # Formata as informações do endereço
            endereco_formatado = f"Endereço para o CEP {cep}:\n"
            endereco_formatado += f"  Logradouro: {data.get('logradouro', 'N/A')}\n"
            endereco_formatado += f"  Complemento: {data.get('complemento', 'N/A')}\n"
            endereco_formatado += f"  Bairro: {data.get('bairro', 'N/A')}\n"
            endereco_formatado += f"  Localidade: {data.get('localidade', 'N/A')}\n"
            endereco_formatado += f"  UF: {data.get('uf', 'N/A')}\n"
            endereco_formatado += f"  DDD: {data.get('ddd', 'N/A')}\n"
            return endereco_formatado

    except requests.exceptions.RequestException as e:
        return f"Erro de conexão com a API ViaCEP: {e}"
    except json.JSONDecodeError:
        return "Erro: Não foi possível decodificar a resposta da API como JSON."
    except Exception as e:
        return f"Ocorreu um erro inesperado: {e}"

# --- Exemplos de uso do script ---

# Exemplo 1: CEP válido
print(consultar_cep("01001-000"))
print("\n" + "-"*30 + "\n")

# Exemplo 2: CEP inválido (formato)
print(consultar_cep("123"))
print("\n" + "-"*30 + "\n")

# Exemplo 3: CEP não existente
print(consultar_cep("99999999"))
print("\n" + "-"*30 + "\n")

# Exemplo 4: Outro CEP válido
print(consultar_cep("70002900"))
```

```python
import time

def enviar_email_simulado(destinatario, assunto, corpo):
    """
    Simula o envio de um e-mail.
    Na vida real, você usaria uma biblioteca como `smtplib`.
    """
    print(f"--- E-MAIL SIMULADO ENVIADO ---")
    print(f"Para: {destinatario}")
    print(f"Assunto: {assunto}")
    print(f"Corpo:\n{corpo}")
    print(f"------------------------------")

def checar_condicao_critica(valor_monitorado):
    """
    Esta função define a lógica para uma condição crítica.
    No exemplo, a condição crítica é atingida se o valor_monitorado for maior que 90.
    """
    limite_critico = 90
    if valor_monitorado > limite_critico:
        return True, f"VALOR CRÍTICO ATINGIDO: {valor_monitorado} excedeu o limite de {limite_critico}!"
    else:
        return False, f"Valor atual: {valor_monitorado}. Tudo OK."

def _disparar_alerta(mensagem_critica, valor_monitorado_original):
    """
    Dispara um alerta no console e envia um e-mail simulado com a mensagem crítica.
    """
    print(f"*** ALERTA NO CONSOLE: {mensagem_critica} ***")
    assunto_email = "ALERTA CRÍTICO: Condição de sistema atingida!"
    corpo_email = (
        f"Prezado administrador,\n\nA seguinte condição crítica foi detectada:\n"
        f"{mensagem_critica}\n"
        f"Valor monitorado no momento do alerta: {valor_monitorado_original}\n\n"
        f"Por favor, tome as ações necessárias.\n\n"
        f"Atenciosamente,\nSeu Sistema de Monitoramento"
    )
    enviar_email_simulado("admin@example.com", assunto_email, corpo_email)

def monitorar_e_alertar():
    print("Iniciando monitoramento...")
    # Simulação de um valor que muda ao longo do tempo
    valores_exemplo = [75, 80, 85, 91, 88, 95, 70]

    for i, valor in enumerate(valores_exemplo):
        print(f"\nIteração {i+1}: Monitorando valor {valor}...")
        condicao_critica_atingida, mensagem = checar_condicao_critica(valor)

        if condicao_critica_atingida:
            _disparar_alerta(mensagem, valor) # Chama a nova função auxiliar
        else:
            print(f"Status: {mensagem}")
        time.sleep(1) # Aguarda 1 segundo para simular o tempo entre verificações

    print("\nMonitoramento concluído.")

# --- Executar o programa de monitoramento ---
monitorar_e_alertar()
```

## Consultar Endereço por CEP com ViaCEP

Este script Python permite consultar informações de endereço utilizando a API pública do ViaCEP. Ele solicita um CEP e retorna o endereço formatado de maneira legível.

```python
import requests

def consultar_cep(cep):
    """
    Consulta a API do ViaCEP para obter informações de endereço.

    Args:
        cep (str): O Código de Endereçamento Postal (CEP) a ser consultado.

    Returns:
        dict or None: Um dicionário com as informações do endereço formatadas,
                      ou None se o CEP for inválido ou ocorrer um erro.
    """
    url = f"https://viacep.com.br/ws/{cep}/json/"
    try:
        response = requests.get(url)
        response.raise_for_status() # Lança um erro para status HTTP ruins (4xx ou 5xx)
        data = response.json()

        if 'erro' in data:
            print(f"CEP {cep} não encontrado ou inválido.")
            return None

        endereco_formatado = {
            "CEP": data.get('cep'),
            "Logradouro": data.get('logradouro'),
            "Complemento": data.get('complemento'),
            "Bairro": data.get('bairro'),
            "Localidade": data.get('localidade'),
            "UF": data.get('uf'),
            "DDD": data.get('ddd')
        }
        return endereco_formatado

    except requests.exceptions.RequestException as e:
        print(f"Ocorreu um erro ao consultar o CEP: {e}")
        return None
    except ValueError:
        print("Erro: A resposta da API não é um JSON válido.")
        return None


# Exemplo de uso:
print("--- Consulta de Endereço ViaCEP ---")
while True:
    cep_input = input("Digite um CEP (apenas números, ou 'sair' para encerrar): ").strip()
    if cep_input.lower() == 'sair':
        break

    # Remove caracteres não numéricos
    cep_limpo = ''.join(filter(str.isdigit, cep_input))

    if len(cep_limpo) != 8:
        print("CEP inválido. O CEP deve conter 8 dígitos numéricos.")
        continue

    endereco = consultar_cep(cep_limpo)

    if endereco:
        print("\n--- Detalhes do Endereço ---")
        for chave, valor in endereco.items():
            if valor:
                print(f"{chave}: {valor}")
        print("---------------------------\n")
    else:
        print(f"Não foi possível obter informações para o CEP: {cep_input}\n")

print("Consulta encerrada.")
```

## Organizador de Arquivos por Extensão

Este script Python organiza arquivos em um diretório especificado, movendo-os para subpastas criadas com base na extensão de cada arquivo (ex: `.pdf`, `.jpg`).

```python
import os
import shutil

def organizar_arquivos_por_extensao(diretorio_origem):
    """
    Organiza arquivos em um diretório, movendo-os para subpastas
    criadas com base na extensão do arquivo.

    Args:
        diretorio_origem (str): O caminho para o diretório que contém os arquivos a serem organizados.
    """
    if not os.path.isdir(diretorio_origem):
        print(f"Erro: O diretório '{diretorio_origem}' não existe.")
        return

    print(f"Iniciando organização no diretório: {diretorio_origem}")

    for filename in os.listdir(diretorio_origem):
        # Ignorar diretórios, apenas processar arquivos
        if os.path.isfile(os.path.join(diretorio_origem, filename)):
            file_path = os.path.join(diretorio_origem, filename)
            file_name, file_extension = os.path.splitext(filename)

            # Se o arquivo não tiver extensão, considerar como 'SEM_EXTENSAO'
            if not file_extension:
                folder_name = 'SEM_EXTENSAO'
            else:
                # Remover o ponto da extensão para usar como nome da pasta
                folder_name = file_extension[1:].upper()

            destination_folder = os.path.join(diretorio_origem, folder_name)

            # Criar a pasta de destino se ela não existir
            if not os.path.exists(destination_folder):
                os.makedirs(destination_folder)
                print(f"Pasta '{folder_name}' criada.")

            # Mover o arquivo para a pasta de destino
            try:
                shutil.move(file_path, destination_folder)
                print(f"Movido '{filename}' para '{destination_folder}'")
            except shutil.Error as e:
                print(f"Erro ao mover o arquivo '{filename}': {e}")
            except Exception as e:
                print(f"Ocorreu um erro inesperado ao mover '{filename}': {e}")

    print("Organização concluída!")

# --- Exemplo de Uso --- #
# Crie um diretório de teste e alguns arquivos para demonstrar
# (Você pode pular esta parte e usar um diretório existente, mas tenha cuidado!)

# Descomente e modifique o caminho abaixo para o seu diretório de interesse.
# Por exemplo, se você quiser organizar arquivos na pasta 'Downloads', use:
# diretorio_a_organizar = '/home/seu_usuario/Downloads'

# ATENÇÃO: Tenha certeza do diretório que você está usando, pois o script
# moverá os arquivos permanentemente!

# Para demonstração no Colab, criaremos um diretório e alguns arquivos temporários.
diretorio_exemplo = './arquivos_baguncados'

# Criar o diretório de exemplo se não existir
if not os.path.exists(diretorio_exemplo):
    os.makedirs(diretorio_exemplo)

# Criar alguns arquivos de exemplo (apenas para demonstração)
open(os.path.join(diretorio_exemplo, 'documento1.pdf'), 'w').close()
open(os.path.join(diretorio_exemplo, 'imagem1.jpg'), 'w').close()
open(os.path.join(diretorio_exemplo, 'relatorio.docx'), 'w').close()
open(os.path.join(diretorio_exemplo, 'foto.png'), 'w').close()
open(os.path.join(diretorio_exemplo, 'planilha.xlsx'), 'w').close()
open(os.path.join(diretorio_exemplo, 'sem_extensao'), 'w').close()
open(os.path.join(diretorio_exemplo, 'outro_documento.pdf'), 'w').close()

print(f"Arquivos de exemplo criados em: {diretorio_exemplo}")
print("Conteúdo antes da organização:")
print(os.listdir(diretorio_exemplo))

# Chamar a função para organizar
organizar_arquivos_por_extensao(diretorio_exemplo)

print("\nConteúdo após a organização:")
print(os.listdir(diretorio_exemplo))

# Para limpar os arquivos de exemplo após o teste (opcional)
# shutil.rmtree(diretorio_exemplo)
# print(f"Diretório de exemplo '{diretorio_exemplo}' removido.")
```
